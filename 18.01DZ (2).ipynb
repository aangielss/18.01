```python
a1 = pd.concat([df5, df12], ignore_index=True)
a2 = pd.concat([df6, df10], ignore_index=True)
a3 = pd.concat([df2, df11], ignore_index=True)
a4 = pd.concat([df1, df9], ignore_index=True)
a5 = pd.concat([df3, df8], ignore_index=True)
a6 = pd.concat([df4, df7], ignore_index=True)
```

```python
df_all = a1.merge(a2, left_on='ID', right_on='ID')
df_all = df_all.merge(a3, left_on='ID', right_on='ID')
df_all = df_all.merge(a4, left_on='ID', right_on='ID')
df_all = df_all.merge(a5, left_on='ID', right_on='ID')
df_all = df_all.merge(a6, left_on='ID', right_on='ID')
df_all = df_all.sort_values(by='ID')
df_all
```

```python
df_all.APP_CAR  = df_all.APP_CAR .str.lower()
df_all.APP_COMP_TYPE  = df_all.APP_COMP_TYPE .str.lower()
df_all.APP_DRIVING_LICENSE  = df_all.APP_DRIVING_LICENSE .str.lower()
df_all.APP_EDUCATION  = df_all.APP_EDUCATION .str.lower()
df_all.APP_EMP_TYPE  = df_all.APP_EMP_TYPE .str.lower()
df_all.APP_KIND_OF_PROP_HABITATION  = df_all.APP_KIND_OF_PROP_HABITATION .str.lower()
df_all.APP_MARITAL_STATUS  = df_all.APP_MARITAL_STATUS .str.lower()
df_all.APP_POSITION_TYPE  = df_all.APP_POSITION_TYPE .str.lower()
df_all.APP_TRAVEL_PASS  = df_all.APP_TRAVEL_PASS .str.lower()
df_all.CLNT_JOB_POSITION  = df_all.CLNT_JOB_POSITION .str.lower()
df_all.CLNT_JOB_POSITION_TYPE  = df_all.CLNT_JOB_POSITION_TYPE .str.lower()
df_all.CLNT_TRUST_RELATION  = df_all.CLNT_TRUST_RELATION .str.lower()
df_all.PACK  = df_all.PACK .str.lower()
```

```python
list(df_all.CLNT_TRUST_RELATION.unique())
```

```python
df_all.CLNT_TRUST_RELATION.replace('mother','мать',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('brother','брат',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('friend','друг',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('sister','сестра',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('other','другое',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('relative','родственник',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('daughter','дочь',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('son','сын',inplace=True)
df_all.CLNT_TRUST_RELATION.replace('father','отец',inplace=True)
```

```python
df_all.APP_CAR.where(df_all.APP_CAR.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_COMP_TYPE.where(df_all.APP_COMP_TYPE.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_DRIVING_LICENSE.where(df_all.APP_DRIVING_LICENSE.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_EDUCATION.where(df_all.APP_EDUCATION.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_EMP_TYPE.where(df_all.APP_EMP_TYPE.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_KIND_OF_PROP_HABITATION.where(df_all.APP_KIND_OF_PROP_HABITATION.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_MARITAL_STATUS.where(df_all.APP_MARITAL_STATUS.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_POSITION_TYPE.where(df_all.APP_POSITION_TYPE.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.APP_TRAVEL_PASS.where(df_all.APP_TRAVEL_PASS.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.CLNT_JOB_POSITION.where(df_all.CLNT_JOB_POSITION.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.CLNT_JOB_POSITION_TYPE.where(df_all.CLNT_JOB_POSITION_TYPE.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
df_all.CLNT_TRUST_RELATION.where(df_all.CLNT_TRUST_RELATION.str.replace(' ','').str.isalpha(), np.nan, inplace=True)
```
